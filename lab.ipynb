# Cardiovascular Disease Prediction

## Introduction

The World Health Organization (WHO) estimates that 17.9 million people die from cardiovascular diseases (CVDs) every year.  
**L'Organisation mondiale de la santé (OMS)** estime que 17,9 millions de personnes meurent chaque année de maladies cardiovasculaires (MCV).

There are multiple risk factors that could contribute to CVD in an individual, such as unhealthy diet, lack of physical activity, or mental illness. Identifying these risk factors early on could help prevent many premature deaths.  
**Il existe de multiples facteurs de risque** qui pourraient contribuer aux maladies cardiovasculaires chez un individu, comme une mauvaise alimentation, le manque d'activité physique ou une maladie mentale. L’identification précoce de ces facteurs de risque pourrait contribuer à prévenir de nombreux décès prématurés.

We are working for an R&D company that focuses on providing healthcare solutions. The company has collected anonymized data from multiple hospitals on several patients. The dataset includes relevant information for each patient, such as their personal information and some medical data, including whether or not they have had heart disease before.  
**Nous travaillons pour une entreprise de R&D** qui se concentre sur la fourniture de solutions de santé. La société a collecté des données anonymisées auprès de plusieurs hôpitaux sur plusieurs patients. L'ensemble de données comprend des informations pertinentes pour chaque patient, telles que ses informations personnelles et certaines données médicales, notamment s'il a déjà eu ou non une maladie cardiaque.

We are tasked with using the dataset to accurately predict the likelihood of a new patient having heart disease in the future. The dataset has the following features:  
**Nous sommes chargés d’utiliser l’ensemble de données** pour prédire avec précision la probabilité qu’un nouveau patient souffre d’une maladie cardiaque à l’avenir. L'ensemble de données présente les caractéristiques suivantes :

## Dataset Features

- **Age**: age of the patient [years]  
  **Age** : âge du patient [années]
  
- **Sex**: sex of the patient [M: Male, F: Female]  
  **Sex** : sexe du patient [M : Homme, F : Femme]
  
- **ChestPainType**: chest pain type [TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic]  
  **ChestPainType** : type de douleur thoracique [TA : angine typique, ATA : angine atypique, NAP : douleur non angineuse, ASY : asymptomatique]
  
- **RestingBP**: resting blood pressure [mm Hg]  
  **RestingBP** : tension artérielle au repos [mm Hg]
  
- **Cholesterol**: serum cholesterol [mm/dl]  
  **Cholesterol** : cholestérol sérique [mm/dl]
  
- **FastingBS**: fasting blood sugar [1: if FastingBS > 120 mg/dl, 0: otherwise]  
  **FastingBS** : glycémie à jeun [1 : si FastingBS > 120 mg/dl, 0 : sinon]
  
- **RestingECG**: resting electrocardiogram results [Normal: Normal, ST: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV), LVH: showing probable or definite left ventricular hypertrophy by Estes' criteria]  
  **RestingECG** : résultats de l'électrocardiogramme au repos [Normal : normal, ST : présentant une anomalie de l'onde ST-T (inversions de l'onde T et/ou élévation ou dépression ST > 0,05 mV), HVG : montrant une anomalie ventriculaire gauche probable ou certaine hypertrophie selon les critères d'Estes]
  
- **MaxHR**: maximum heart rate achieved [Numeric value between 60 and 202]  
  **MaxHR** : fréquence cardiaque maximale atteinte [Valeur numérique comprise entre 60 et 202]
  
- **ExerciseAngina**: exercise-induced angina [Y: Yes, N: No]  
  **ExerciseAngina** : angine induite par l'exercice [Y : Oui, N : Non]
  
- **Oldpeak**: oldpeak = ST [Numeric value measured in depression]  
  **Oldpeak** : oldpeak = ST [Valeur numérique mesurée en dépression]
  
- **ST_Slope**: the slope of the peak exercise ST segment [Up: upsloping, Flat: flat, Down: downsloping]  
  **ST_Slope** : la pente du segment ST de l'exercice de pointe [Up : ascendant, Flat : plat, Down : descendant]
  
- **HeartDisease**: output class [1: heart disease, 0: Normal]  
  **HeartDisease** : classe de sortie [1 : maladie cardiaque, 0 : Normal]

Remplacer par la médiane

```python
median_values = df_original.groupby('HeartDisease')[['RestingBP', 'Cholesterol']].median()
```

```python
median_values
```

```python
 
df = df_original.copy()

# Supprimer les valeurs invalides de RestingBP
df = df[df["RestingBP"] != 0]

# Calculer les médianes de Cholesterol pour chaque groupe HeartDisease
medians = df.groupby("HeartDisease")["Cholesterol"].median()

# Remplacer les valeurs 0 dans Cholesterol par la médiane du groupe correspondant
for heart_disease_value, median_value in medians.items():
    # Identifier les lignes où Cholesterol est 0 et où HeartDisease correspond au groupe
    condition = (df["HeartDisease"] == heart_disease_value) & (df["Cholesterol"] == 0)
    df.loc[condition, "Cholesterol"] = median_value
```

Créer une carte thermique de corrélation de Pearson:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculer la matrice de corrélation
correlation_matrix = df.corr()

# Tracer une carte thermique
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Carte thermique de corrélation de Pearson")
plt.show()
```

```python
filtered_features = correlation_matrix['HeartDisease'][
    (correlation_matrix['HeartDisease'] > 0.3) | (correlation_matrix['HeartDisease'] < -0.3)
]
# Étape 3 : Afficher les résultats
print("Caractéristiques avec une corrélation supérieure à 0.3 avec 'HeartDisease':")
print(filtered_features )
```

```python
# Sélectionner les caractéristiques en fonction de r^2
correlation_with_target = correlation_matrix['HeartDisease']
correlation_squared = correlation_with_target ** 2
print("Carré du coefficient de corrélation avec HeartDisease :")
print(correlation_squared.sort_values(ascending=False))
```

 
### **1. Caractéristiques les plus importantes (\(r^2 > 0.1\))**
Ces caractéristiques sont les plus pertinentes pour prédire `HeartDisease` :
- **`ST_Slope_Up` (\(r^2 = 0.387\))** : Une pente montante sur le test ECG est fortement corrélée avec une absence de maladie cardiaque.
- **`ST_Slope_Flat` (\(r^2 = 0.307\))** : Une pente plate est corrélée à une probabilité accrue de maladie cardiaque.
- **`ExerciseAngina_Y` (\(r^2 = 0.244\))** : Angine induite par l'effort physique est un indicateur clé de maladie cardiaque.
- **`Oldpeak` (\(r^2 = 0.163\))** : Dépression du segment ST lors d'un test de stress est significative.
- **`ChestPainType_ATA` (\(r^2 = 0.162\))** : L'angine atypique est un facteur important.
- **`MaxHR` (\(r^2 = 0.160\))** : La fréquence cardiaque maximale est légèrement corrélée.

Ces caractéristiques devraient être priorisées dans le modèle.

---

### **2. Caractéristiques modérément importantes (\(0.05 < r^2 \leq 0.1\))**
- **`Sex_M` (\(r^2 = 0.093\))** : Le sexe masculin a une influence, mais moins significative.
- **`Age` (\(r^2 = 0.080\))** : L'âge contribue, mais pas de manière dominante.
- **`FastingBS` (\(r^2 = 0.071\))** : La glycémie à jeun élevée a une faible corrélation.

Ces caractéristiques peuvent être incluses pour améliorer légèrement la performance du modèle.

---

### **3. Caractéristiques moins pertinentes (\(r^2 \leq 0.05\))**
- **`ChestPainType_NAP` (\(r^2 = 0.045\))** : Angine non typique a une faible corrélation.
- **`RestingBP` (\(r^2 = 0.014\))** : La tension artérielle au repos n'est presque pas corrélée.

## 1. Rapport de Classification

Le rapport de classification fournit plusieurs métriques importantes pour évaluer la performance de votre modèle, en particulier la **précision (precision)**, le **rappel (recall)**, et le **score F1 (f1-score)** pour chaque classe, ainsi que l'**exactitude (accuracy)** globale.

- **Classe 0 (pas de maladie cardiaque)** : 0.92
- **Classe 1 (maladie cardiaque)** : 0.88

La précision mesure la proportion des prédictions positives correctes parmi toutes les prédictions positives faites.  
  - Pour la classe "pas de maladie cardiaque" (classe 0), la précision de 92% signifie que parmi les cas prédits comme "pas de maladie cardiaque", 92% étaient effectivement corrects.
  - Pour la classe "maladie cardiaque" (classe 1), la précision est de 88%, ce qui est également assez bon, mais un peu inférieur à la classe 0.

### Rappel (Recall)
- **Classe 0** : 0.87
- **Classe 1** : 0.93

Le rappel mesure la capacité du modèle à identifier correctement les instances positives parmi toutes les instances réelles positives.  
  - La classe "pas de maladie cardiaque" a un rappel de 87%, ce qui signifie que 87% des patients sans maladie cardiaque ont été correctement classés comme tels.
  - La classe "maladie cardiaque" a un rappel de 93%, ce qui montre que le modèle est excellent pour identifier les patients atteints de maladies cardiaques (moins de faux négatifs).

### Score F1
- **Classe 0** : 0.89
- **Classe 1** : 0.90

Le **score F1** est la moyenne harmonique de la précision et du rappel. Il équilibre la précision et le rappel. Un F1-score de 0.89 pour la classe "pas de maladie cardiaque" et 0.90 pour la classe "maladie cardiaque" indique que le modèle a un bon équilibre entre la capacité à éviter les faux positifs et les faux négatifs pour les deux classes.

- **0.90** (ou 90%)

Le score moyen de la précision, du rappel et du score F1 sur toutes les classes est de 0.90, ce qui indique que le modèle est globalement bien équilibré entre les classes.

### Moyenne pondérée
Les valeurs de la moyenne pondérée (pour la précision, le rappel et le score F1) sont également de 0.90, ce qui prend en compte le nombre d'échantillons dans chaque classe. Cela montre que le modèle est bien généraliste et fonctionne également bien pour les classes déséquilibrées.
